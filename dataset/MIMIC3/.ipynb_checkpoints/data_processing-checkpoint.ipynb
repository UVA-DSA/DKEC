```python
train = pd.read_csv('clean_train.csv', index_col=0)
test = pd.read_csv('clean_test.csv', index_col=0)
val = pd.read_csv('clean_val.csv', index_col=0)
```

```python
train_path = '/home/xueren/Downloads/caml-mimic-master/mimicdata/mimic3/train_full.csv'
test_path = '/home/xueren/Downloads/caml-mimic-master/mimicdata/mimic3/train_full.csv'
val_path = '/home/xueren/Downloads/caml-mimic-master/mimicdata/mimic3/dev_full.csv'

raw_train = pd.read_csv(train_path)
raw_test = pd.read_csv(test_path)
raw_val = pd.read_csv(val_path)
```

```python
SUBJECT_ID = raw_train['SUBJECT_ID'].tolist()

for i in range(len(train)):
    index = SUBJECT_ID.index(train['SUBJECT_ID'][i])
    if train['SUBJECT_ID'][i] == raw_train['SUBJECT_ID'][index] and train['HADM_ID'][i] == raw_train['HADM_ID'][index]:
        train['TEXT'][i] = raw_train['TEXT'][i]
        train['length'][i] = raw_train['length'][i]
        cur_label = []
        for l in raw_train['LABELS'][i].split(';'):
            if l.strip() in label:
                cur_label.append(l.strip())
        train['LABELS'][i] = ';'.join(cur_label)
    else:
        raise Exception('error')
```
